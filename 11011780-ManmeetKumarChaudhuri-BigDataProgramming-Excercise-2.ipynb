```python
import random
import time

#create cards for the game 
class Card:
    def __init__(self, name, rank, height, weight, chest, biceps):
        self.name = name
        self.rank = rank
        self.height = height
        self.weight = weight
        self.chest = chest
        self.biceps = biceps
        
cards = []
cards.append(Card("Jack Snake", 1, 1.8, 120, 46, 19)) 
cards.append(Card("Hulk Hogan", 2, 1.9, 130, 48, 21))
cards.append(Card("Randy Orten", 3, 1.75, 140, 47, 22))
cards.append(Card("Matt Hardy", 4, 1.85, 115, 50, 20))
cards.append(Card("Hit Man", 5, 1.78, 151, 51, 23))
cards.append(Card("Under Taker", 6, 1.89, 100, 52, 24))
cards.append(Card("Andre Gaint", 7, 1.93, 170, 53, 18))
cards.append(Card("Harry Styles", 8, 1.81, 175, 54, 16))
cards.append(Card("Owen Hart", 9, 1.82, 161, 55, 17))
cards.append(Card("shawn Micheals", 10, 1.92, 200, 56, 26))
cards.append(Card("Steve Austin", 11, 1.98, 104, 57, 25))
cards.append(Card("Berry More", 12, 1.65, 152, 58, 27))
cards.append(Card("Rey Misterio", 13, 1.60, 149, 59, 28))
cards.append(Card("Razor Ramon", 1.55, 159, 200, 60, 29))
cards.append(Card("The Batista", 15, 1.88, 179, 61, 30))
cards.append(Card("John Cena", 16, 1.74, 162, 62, 31))

random.shuffle(cards)

playerDeck = []
computerDeck = []
outdatedDeck = []



# Deal the cards
while len(cards) > 0:
    playerDeck.append(cards.pop(0))
    computerDeck.append(cards.pop(0))


#rolling dices
print("TRUMP CARDS GAME: Player1 vs Player2!")
game_start = input("Players, would you like to roll the dice? yes or no ")
def dice_roll():
    x1=random.randint(1,6)
    x2=random.randint(1,6)
    global playerTurn
    print("Player1 dice number is: " + str(x1))
    print("Player2 dice number is: " + str(x2))  
    if x1>x2:
        print("Player1 calls first!")
        playerTurn = True
    elif x1<x2:
        print("Player2 calls first!")
        playerTurn = False
    else:
        print("It's a tie. Roll again!")
        dice_roll()
    
if game_start == "yes":
    dice_roll()
elif game_start == "no":
    print("Please input yes to proceed")
else:
    print("Input not recognized")
    

#start the game    
playerTurn = True
allowedResponses = ["a", "b", "c", "d", "e"]

    
gspellcountp = 0
gspellcountc = 0
rspellcountp = 0
rspellcountc = 0
p1 = 0
p2 = 0
rspell = "no"
rspell1 = "no"

while len(playerDeck) > 0 and len(computerDeck) > 0:
    
    time.sleep(1)

    if playerTurn == True:
        #player1 resurrect spell
        ra = 0
        if (rspellcountp == 0 and len(outdatedDeck) > 1 and ra == 0):
            rspell = input("Player1, do you want to play Resurrect Spell? yes or no: ")
            if (rspell == "yes" and rspellcountp == 0):
                z1 = random.randint(1,len(outdatedDeck))
                playerCard = outdatedDeck.pop(int(z1)-1)
                outdatedDeck.append(playerCard)
                rspellcountp = 1
                ra = 1
            else:
                playerCard = playerDeck.pop(0)
                outdatedDeck.append(playerCard)

        else:
            playerCard = playerDeck.pop(0)
            outdatedDeck.append(playerCard)

            
        print("PLAYER1 CARD")
        print("Name:", playerCard.name)
        print("a. Rank:", playerCard.rank)
        print("b. Height:", playerCard.height)
        print("c. Weight:", playerCard.weight)
        print("d. Chest:", playerCard.chest)
        print("e. Biceps:", playerCard.biceps)
        answer = input("Player1, which property do you choose? ")
        
        while allowedResponses.count(answer) == 0:
            answer = input("That isn't a valid answer, please try again: ")
        #player1 god spell
        if (gspellcountp == 0  and len(computerDeck) > 1 and ra == 0):
            gspell = input("Player1, do you want to play God Spell? yes or no: ")
            if (gspell == "yes" and gspellcountp == 0):
                gspellcountp = 1
                lec = len(computerDeck)
                print("No of cards in Player2 deck:", lec)
                cardno = input("Which card number from Player2 deck? ")
                if (rspellcountc == 0 and len(outdatedDeck) > 1):
                    rspell1 = input("Player2, do you want to play Resurrect Spell? yes or no: ")
                    if (rspell1 == "yes" and rspellcountc == 0):
                        z2 = random.randint(1,len(outdatedDeck))
                        mn = outdatedDeck.pop(int(z2)-1)
                        computerDeck.insert(0,mn)
                        choice = input("Player1: a. Force resurrected card or b. Force earlier choice ")
                        if (choice == "a"):
                            computerCard = computerDeck.pop(0)
                            outdatedDeck.append(computerCard)
                            rspellcountc = 1
                        else:    
                            computerCard = computerDeck.pop(int(cardno)-1)
                            outdatedDeck.append(computerCard)
                            rspellcountc = 1
                    else:
                        computerCard = computerDeck.pop(int(cardno)-1)
                        outdatedDeck.append(computerCard)
                else:
                    computerCard = computerDeck.pop(int(cardno)-1)
                    outdatedDeck.append(computerCard) 
                    
            else:
                computerCard = computerDeck.pop(0)
                outdatedDeck.append(computerCard) 

        
        else:
            computerCard = computerDeck.pop(0)
            outdatedDeck.append(computerCard) 
        
        print("PLAYER2 CARD")
        print("Name:", computerCard.name)
        print("a. Rank:", computerCard.rank)
        print("b. Height:", computerCard.height)
        print("c. Weight:", computerCard.weight)
        print("d. Chest:", computerCard.chest)
        print("e. Biceps:", computerCard.biceps)
     
    
    else:
        #player2 resurrect spell
        rb = 0
        if (rspellcountc == 0 and len(outdatedDeck) > 1 and rb == 0):
            rspell1 = input("Player2, do you want to play Resurrect Spell? yes or no: ")
            if (rspell1 == "yes" and rspellcountc == 0):
                z2 = random.randint(1,len(outdatedDeck))
                computerCard = outdatedDeck.pop(int(z2)-1)
                outdatedDeck.append(computerCard)
                rspellcountc = 1
                rb = 1
            else:
                computerCard = computerDeck.pop(0)
                outdatedDeck.append(computerCard)

        else:
            computerCard = computerDeck.pop(0)
            outdatedDeck.append(computerCard)
        
       
        print("PLAYER2 CARD")
        print("Name:", computerCard.name)
        print("a. Rank:", computerCard.rank)
        print("b. Height:", computerCard.height)
        print("c. Weight:", computerCard.weight)
        print("d. Chest:", computerCard.chest)
        print("e. Biceps:", computerCard.biceps)
        
        answer = input("Player2, which property do you choose? ")
        print("Player2 chooses", answer)
        
        #Player2 god spell
        if (gspellcountc == 0 and len(playerDeck) > 1 and rb == 0):
            gspell1 = input("Player2, do you want to play God Spell? yes or no: ")
            if (gspell1 == "yes" and gspellcountc == 0):
                gspellcountc = 1
                lep = len(playerDeck)
                print("No of cards in Player1 deck:", lep)
                cardno1 = input("Which card number from Player1 deck? ")
                if (rspellcountp == 0 and len(outdatedDeck) > 1):
                    rspell = input("Player1, do you want to play Resurrect Spell? yes or no: ")
                    if (rspell == "yes" and rspellcountp == 0):
                        z1 = random.randint(1,len(outdatedDeck))
                        nm = outdatedDeck.pop(int(z1)-1)
                        playerDeck.insert(0,nm)
                        choice1 = input("Player2: a. Force resurrected card or b. Force earlier choice ")
                        if (choice1 == "a"):
                            playerCard = playerDeck.pop(0)
                            outdatedDeck.append(playerCard)
                            rspellcountp = 1
                        else:
                            playerCard = playerDeck.pop(int(cardno1)-1)
                            outdatedDeck.append(playerCard)
                            rspellcountp = 1
                    else:
                        playerCard = playerDeck.pop(int(cardno1)-1)
                        outdatedDeck.append(playerCard)
                else:                 
                    playerCard = playerDeck.pop(int(cardno1)-1)
                    outdatedDeck.append(playerCard)
            else:
                playerCard = playerDeck.pop(0)
                outdatedDeck.append(playerCard)

        
        else:
            playerCard = playerDeck.pop(0)
            outdatedDeck.append(playerCard)
                                   
        print("PLAYER1 CARD")
        print("Name:", playerCard.name)
        print("a. Rank:", playerCard.rank)
        print("b. Height:", playerCard.height)
        print("c. Weight:", playerCard.weight)
        print("d. Chest:", playerCard.chest)
        print("e. Biceps:", playerCard.biceps)
    
    
    playerWins = False
    #comparing cards of player1 and player2
    if answer == "a":
        playerWins = (playerCard.rank < computerCard.rank)
    elif answer == "b":
        playerWins = (playerCard.height > computerCard.height)
    elif answer == "c":
        playerWins = (playerCard.weight > computerCard.weight)
    elif answer == "d":
        playerWins = (playerCard.chest > computerCard.chest)
    elif answer == "e":
        playerWins = (playerCard.biceps > computerCard.biceps)

    time.sleep(1)
    
    #who wins the hand?
    if playerWins:
        print("Player1 wins this hand!")
        p1=p1+1      
        playerTurn = True
    else:
        print("Player2 wins this hand!")
        p2=p2+1
        playerTurn = False

time.sleep(2)

#who wins the match?
if p1<p2:
    print("PLAYER2 WINS THE MATCH!")
elif p1>p2:
    print("PLAYER1 WINS THE MATCH!")
else:
    print("IT'S A TIE!")
 

time.sleep(2)

#total points of player1 and player2 in the match
print("Player1 total points: ", p1)
print("Player2 total points: ", p2)
print("Total cards in Outdated Deck: ", len(outdatedDeck))

time.sleep(2)
```
